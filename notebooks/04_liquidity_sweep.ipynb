```python
import pandas as pd
from trader.backtest_harness import run_fold

def pooled_trades(df, strategy_cls, initial_train_months=12, test_months=3, cash=25_000):
    data_start, data_end = df.index[0], df.index[-1]
    train_end = data_start + pd.DateOffset(months=initial_train_months)
    all_trades = []
    while train_end + pd.DateOffset(months=test_months) <= data_end:
        test_end = train_end + pd.DateOffset(months=test_months)
        _, test_stats = run_fold(df, data_start, train_end, test_end, cash=cash,
                                  strategy_cls=strategy_cls, optimize_kwargs=None, verbose=False)
        trades = test_stats._trades.copy()
        trades["fold_test_end"] = test_end
        all_trades.append(trades)
        train_end = test_end
    return pd.concat(all_trades, ignore_index=True)

for symbol in ["^DJI", "GC=F"]:
    df = build_kalman_features(build_liquidity_features(build_bt_df(symbol)))
    trades = pooled_trades(df, LiquiditySweepStrategy)
    wins = (trades["PnL"] > 0).sum()
    print(symbol, "trades:", len(trades), "win rate:", wins / len(trades) if len(trades) else float("nan"), "total PnL:", trades["PnL"].sum())
```
